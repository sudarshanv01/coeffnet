```python
from e3nn.o3 import FullTensorProduct, FullyConnectedTensorProduct, Irreps
from e3nn.util.test import equivariance_error
```

```python
# Show the full tensor product
tp = FullTensorProduct(
    irreps_in1='62x0e + 18x1o + 9x2e',
    irreps_in2='62x0e + 18x1o + 9x2e',
)
print(tp)
tp.visualize();
```

```python
# Show the full tensor product
tp = FullyConnectedTensorProduct(
    irreps_in1='62x0e + 18x1o + 9x2e',
    irreps_in2='62x0e + 18x1o + 9x2e',
    irreps_out='1x0e + 1x1o + 1x2e',
)
print(tp)
tp.visualize();
```

```python
irrep_test = Irreps('62x0e + 18x1o + 9x2e')
tp = FullyConnectedTensorProduct(
    irreps_in1=irrep_test,
    irreps_in2=irrep_test,
    irreps_out='1x0e + 1x1o + 1x2e',
)
equivariance_error(
    tp,
    args_in=[tp.irreps_in1.randn(1, -1, 161), tp.irreps_in2.randn(1, -1, 161)],
    irreps_in=[tp.irreps_in1, tp.irreps_in2],
    irreps_out=[tp.irreps_out]
)
```

```python
irrep_test = Irreps('62x0e + 18x1o + 9x2e')
```

```python
from e3nn import o3
rot = -o3.rand_matrix()

D = irrep_test.D_from_matrix(rot)
```

```python

import matplotlib.pyplot as plt
plt.imshow(D, cmap='bwr', vmin=-1, vmax=1);
```
